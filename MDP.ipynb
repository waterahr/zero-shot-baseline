```python
import pickle

# Load image features
file_feature = './results/' + 'Inceptionv3a25DataB_features_all.pickle'
fdata = open(file_feature, 'rb')
features_dict = pickle.load(fdata)  # variables come out in the order you put them in
fdata.close()
features_all = features_dict['features_all']
labels_all = features_dict['labels_all']
images_all = features_dict['images_all']

# Calculate number
train_num = 0 
test_num = 0
#labels_all 
for lab in labels_all:
    if lab[0] != 't':
        train_num += 1
    else:
        test_num += 1
print(train_num, '\t', test_num)
```

```python
import numpy as np

print(features_all.shape)
print(np.shape(labels_all))
print(np.shape(images_all))
print(features_dict)
```

```python
#features_all = features_dict['features_all']
print(np.max(features_all))
print(np.argmax(features_all))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=100)
features_all_pca = pca.fit_transform(features_all)
```

```python
import pandas as pd

f = open(r'/home/anhaoran/data/zero-shot-tianchi/dataset_B/DatasetB_20180919/attributes_per_class.txt')
lines_attr = f.readlines()
attributes = dict()
for each in lines_attr:
    tokens = each.split()
    #print(tokens)
    label = tokens[0]
    attr_list = list()
    for idx in range(1, 26):
        attr_list.append(float(tokens[idx]))
    if not (len(attr_list) == 25):
        print(('attributes number error\n'))
        exit()
    attributes[label] = attr_list
print(attributes)

list_train = list()
list_test = list()
for img_lab in attributes.keys():
    if not img_lab in labels_all:
        list_test.append(img_lab)
    else:
        list_train.append(img_lab)
print(len(list_train))
print(list_train)
print(len(list_test))
print(list_test)
```

```python
# Calculate prototypes (cluster centers)
print(np.max(abs(features_all)))
features_all = features_all/np.max(abs(features_all))
dim_f = features_all.shape[1]
prototypes_train = np.ndarray((len(list_train), dim_f))

dim_a = 25
attributes_train = np.ndarray((len(list_train), dim_a))
attributes_test = np.ndarray((len(list_test), dim_a))

for i in range(len(list_train)):
    label = list_train[i]
    idx = [pos for pos, lab in enumerate(labels_all) if lab == label]
    prototypes_train[i, :] = np.mean(features_all[idx, :], axis=0)
    attributes_train[i, :] = np.asarray(attributes[label])

for i in range(len(list_test)):
    label = list_test[i]
    attributes_test[i, :] = np.asarray(attributes[label])
```

```python
print(prototypes_train.shape)
print(prototypes_train)
print(attributes_train.shape)
print(attributes_train)
print(attributes_test.shape)
print(attributes_test)
```

```python
# Structure learning
LASSO = models.Lasso(alpha=0.01)
LASSO.fit(attributes_train.transpose(), attributes_test.transpose())
W = LASSO.coef_

# Image prototype synthesis
prototypes_test = (np.dot(prototypes_train.transpose(), W.transpose())).transpose()
```

```python
# Prediction - 1 step
label = 'test'
idx = [pos for pos, lab in enumerate(labels_all) if lab == label]
features_test = features_all[idx, :]
images_test = [images_all[i] for i in idx]
prediction = list()

for i in range(len(idx)):
    temp = np.repeat(np.reshape((features_test[i, :]), (1, dim_f)), len(list_test), axis=0)
    distance = np.sum((temp - prototypes_test)**2, axis=1)
    pos_unseen = np.argmin(distance)
    diatance_min_unseen = np.min(distance)
    temp = np.repeat(np.reshape((features_test[i, :]), (1, dim_f)), len(list_train), axis=0)
    distance = np.sum((temp - prototypes_train)**2, axis=1)
    pos_seen = np.argmin(distance)
    diatance_min_seen = np.min(distance)
    if diatance_min_seen < diatance_min_unseen:
        prediction.append(list_train[pos_seen])
    else:
        prediction.append(list_test[pos_unseen])
```

```python
import datetime

now_time = datetime.datetime.now().strftime('%Y-%m-%d')

txtName = "./results/submission_" + now_time + "_inceptionv30910-1step-min.txt"
f=open(txtName, "w")
for i in range(len(prediction)):
    new_context = images_all[train_num + i] +'\t' + prediction[i] + '\n'
    f.write(new_context)
f.close()

sum_unseen = 0
for pre in prediction:
    print(pre)
    if pre in list_test:
        sum_unseen += 1
print(sum_unseen)
print(sum_unseen * 1.0 / len(prediction))
```

```python
# Prediction - 2 steps
label = 'test'
idx = [pos for pos, lab in enumerate(labels_all) if lab == label]
#features_test = features_all_pca[idx, :]
features_test = features_all[idx, :]
images_test = [images_all[i] for i in idx]
prediction = list()

distance_min_unseen = 0
distance_avg_unseen = 0
for i in range(len(idx)):
    temp = np.repeat(np.reshape((features_test[i, :]), (1, dim_f)), len(list_test), axis=0)
    distance = np.sum((temp - prototypes_test)**2, axis=1)
    pos = np.argmin(distance)
    if i == 0 or np.min(distance) < distance_min_unseen:
        distance_min_unseen = np.min(distance)
    distance_avg_unseen += np.min(distance)
    prediction.append(list_test[pos])
distance_avg_unseen /= len(idx)
```
