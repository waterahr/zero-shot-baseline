```python
import os
import random
import shutil

test_spilt = 0.2

train_dir = "/home/anhaoran/data/zero-shot-tianchi/dataset_Ac/train/"
val_dir = "/home/anhaoran/data/zero-shot-tianchi/dataset_Ac/val/"
#train_dir = "./train/train/"
#val_dir = "./train/val/"
if not os.path.exists(val_dir):
    os.mkdir(val_dir)
else:
    print(val_dir, " has been Exist.")
if not os.path.exists(train_dir):
    os.mkdir(train_dir)
else:
    print(train_dir, " has been Exist.")
```

```python
for dirpath, dirnames, filenames in os.walk(train_dir):
    print("current dir: ", dirpath)
    dirname = dirpath[dirpath.rindex('/')+1:]
    new_dirname = os.path.join(val_dir, dirname)
    if not dirname.startswith('.') and not os.path.exists(new_dirname):
        os.mkdir(new_dirname)
        print("make dir: ", new_dirname)
    random.shuffle(filenames)
    number = int(len(filenames) * test_spilt)
    for filename in filenames[:number]:
        src_file = os.path.join(dirpath, filename)
        dst_file = os.path.join(new_dirname, filename)
        shutil.move(src_file, dst_file)
        print(src_file, "--->", dst_file)
```

```python
train_sum = 0
for dirpath, dirnames, filenames in os.walk(train_dir):
    for filename in filenames:
        train_sum += 1
print(train_sum)

test_sum = 0
for dirpath, dirnames, filenames in os.walk(val_dir):
    for filename in filenames:
        test_sum += 1
print(test_sum)

print(train_sum + test_sum)
```
